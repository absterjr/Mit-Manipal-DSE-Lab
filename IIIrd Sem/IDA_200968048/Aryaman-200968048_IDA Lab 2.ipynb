```python
import pandas as pd
```

```python
df = pd.read_csv("bollywood.csv")
```

```python
df
```

```python
#Q1 : Records in the database
df.shape
```

```python
#Q2 : Sort in descending order
df["ReleaseTime"].value_counts()
```

```python
#Q3 : Highest Genre Sales
df[df['ReleaseTime'] == 'FS']['Genre'].value_counts()
```

```python
#Q4 : Crosstab between Genre and Release Time
pd.crosstab(df.Genre,df.ReleaseTime)
```

```python
df['Release Date'] = pd.to_datetime(df['Release Date'])
df['Year'] = df['Release Date'].dt.year
df
#Q5 Extracting a new column called Year

```

```python
#Q5 : Max sales in which year
df["Year"].value_counts().idxmax()
```

```python
#Q6 : Creating month column
df['month'] = pd.DatetimeIndex(df['Release Date']).month
df
```

```python
#Q6 : Budget of 30 crores or more
df[df['Budget'] >= 30]['month'].value_counts().idxmax()
```

```python
#Q6
print("2 is February")
```

```python
#Q7 : ROI
df['ROI']=(df['BoxOfficeCollection']-df['Budget'])/df['Budget']
df.head()
df['ROI']=(df['BoxOfficeCollection']-df['Budget'])/df['Budget']
df.head()
```

```python
#Q7: Flop movies
minimumreturn=df[['MovieName','ROI']].sort_values(by='ROI',ascending=True)
minimumreturn.head(10)
```

```python
#Q8 : Average ROI
df.groupby("ReleaseTime")['Budget'].mean()
```

```python
#Q9 : Correlation
correlation = df[['BoxOfficeCollection' , "YoutubeLikes"]].corr()
print(correlation)
```

```python
#Q10 : More Youtube views
df.boxplot(by="Genre" , column=['YoutubeViews'])

```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#Q11 : Highly Correlated
cordf=df.corr()
sns.heatmap(cordf, annot=True)



```

```python
#Q12 : Barplot
barplot=df.groupby('Genre')['BoxOfficeCollection'].sum()
barplot.plot.bar(color ='black',
width =0.8)
plt.title(" genre of movies and their box office collection")
plt.xlabel('Genre')
plt.ylabel('BoxOfficeCollection')
plt.show()

```

```python
#Q13 : No of movies releases
release = df.groupby("Year")["SlNo"].count()
release
```

```python
#Q13 : Bar plot
release.plot.bar(color="black")
plt.title("Movies released in a year")
plt.xlabel("Year")
plt.ylabel("No. Of Movies")
plt.show()
```

```python
#Q14 : Genre budget
df[["Budget","Genre"]]
```

```python
#Q14 : Plot
barplot=df.groupby('Genre')["Budget"].sum()
barplot.plot.bar(width=0.5)
barplot.plot.bar(color="black")
plt.title(" Budget for every Genre")
plt.xlabel('Genre')
plt.ylabel('Budget ')
plt.show()

```

```python
#Q15 : No of likes and dislikes
df[["YoutubeLikes","YoutubeDislikes","Year"]]
```

```python
youtube=df.groupby('Year')["YoutubeLikes","YoutubeDislikes"].sum()
youtube.plot.bar(width=0.5)

plt.title(" YoutubeLikes and YoutubeDislikes for every year")
plt.xlabel('Year')
plt.ylabel("YoutubeLikes and YoutubeDislikes")
plt.show()
```
